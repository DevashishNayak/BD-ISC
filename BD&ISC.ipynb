```python
#Playfair Cipher
c=[]
t=[]
k=input("Enter Key").upper()
l=len(k)
k=k.replace('J','I')

def removeDuplicate(k, l):
    index = 0
    for i in range(0, l):
        for j in range(0, i + 1): 
            if (k[i] == k[j]):
                break
        if (j == i): 
            k[index] = k[i] 
            index += 1
    return "".join(k[:index])
k1=removeDuplicate(list(k), l)

k2=k1+"ABCDEFGHIKLMNOPQRSTUVWXYZ"
l1=len(k2)
k3=removeDuplicate(list(k2), l1)
k4=list(k3)

import numpy as np
key=np.array(k4).reshape(5,5)
print(key)

p=input("Enter Plain Text").lower()
p=p.replace('j','i')
p1=p.upper()
l2=len(p1)
p2=list(p1)

i=0
while i<len(p2)-1:
    if p2[i]==p2[i+1]:
        p2.insert(i+1,'X')
        i=i+2
    i+=1

l3=len(p2)
if l3%2==1:
    p2.append('X')
pt="".join(p2).lower()
print(pt)

def find_position(letter):
    x=y=0
    for i in range(5):
        for j in range(5):
            if key[i][j]==letter:
                x=i
                y=j
    return x,y

l4=len(p2)
for i in range(0,l4-1,2):
    p3=p2[i]
    p4=p2[i+1]
    x1,y1=find_position(p3)
    x2,y2=find_position(p4)
    if x1==x2:
        if y1==4:
            y1=-1
        if y2==4:
            y2=-1
        c.append(key[x1][y1+1])
        c.append(key[x1][y2+1])
    elif y1==y2:
        if x1==4:
            x1=-1;
        if x2==4:
            x2=-1;
        c.append(key[x1+1][y1])
        c.append(key[x2+1][y2])
    else:
        c.append(key[x1][y2])
        c.append(key[x2][y1])
cipher="".join(c)
print(cipher)

for i in range(0,l4-1,2):
    p3=c[i]
    p4=c[i+1]
    x1,y1=find_position(p3)
    x2,y2=find_position(p4)
    if x1==x2:
        if y1==4:
            y1=-1
        if y2==4:
            y2=-1
        t.append(key[x1][y1-1])
        t.append(key[x1][y2-1])
    elif y1==y2:
        if x1==4:
            x1=-1;
        if x2==4:
            x2=-1;
        t.append(key[x1-1][y1])
        t.append(key[x2-1][y2])
    else:
        t.append(key[x1][y2])
        t.append(key[x2][y1])
    
t1=t.copy()
for i in range(1,l4-1):
    if "X" in t:
        z1=t[i-1]
        z2=t[i+1]
        if z1==z2:
            t1.remove("X")
        
plain="".join(t1).lower()
print(plain)
```

```python
#Galois Field
import numpy as np

n=int(input("Enter n"))
f=2**n
g=[]
h=[]

def d(m):
    d=[1]*(n+n-1)
    for i in range(n+n-1):
        d[i]=d[i]^m[i]
    return d

def m(a,b):
    x=a[::-1]
    y=b[::-1]
    m=[0]*(n+n-1)
    for i in range(len(x)):
        for j in range(len(y)):
            m[i+j]+=int(x[i])*int(y[j])
            if m[i+j]>1:
                m[i+j]=m[i+j]%2
    if m[-1]==1 and len(m)!=1:
        m=d(m)
    l=list(map(str,m[::-1]))
    z="".join(l)
    return z
    
for i in range(f):
    for j in range(f):
        x=bin(i).replace("0b", "")
        y=bin(j).replace("0b", "")
        a=int(x,2)^int(y,2)
        g.append(a)
        z=m(x,y)
        b=(int(z,2))%f
        h.append(b)

a=np.array(g).reshape(f,f)
m=np.array(h).reshape(f,f)
print(a)
print(m)
```

```python
def winner(erica, bob):
    e=0
    b=0
    for i in erica:
        if(i=='E'):
            e+=1
        elif(i=='M'):
            e+=3
        else:
            e+=5
    for i in bob:
        if(i=='E'):
            b+=1
        elif(i=='M'):
            b+=3
        else:
            b+=5
    if(e==b):
        return "tie"
    elif(e>b):
        return "erica"
    else:
        return "bob"
print(winner("E","E"))
```

```python
from mrjob.job import MRJob
from mrjob.step import MRStep

class StatesPopulation(MRJob):
    def mapper_states(self, _, line):
        details = line.split(',')
        yield (details[0]), int(details[2])

    def reducer_sum(self, states, population):
        yield None, (sum(population), states)

    def reducer_top_states(self, _, states_pairs):
        pairs = sorted(states_pairs, reverse=True)
        average = 0
        for i in pairs:
            average+=i[0]
        average/=len(pairs)
            yield pair[0], pair[-1], average

    def steps(self):
        return [
            MRStep(mapper=self.mapper_states,
                reducer=self.reducer_sum),
            MRStep(reducer=self.reducer_top_states)
        ]


if _name_ == '_main_':
    StatesPopulation.run()
```
